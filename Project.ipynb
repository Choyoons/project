```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 데이터 불러오기 
df_vgm2 = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/datasets/vgames2.csv')
df_vgm2
```

```python
index_n = df_vgm2[df_vgm2['NA_Sales'].str.contains('K')].index
index_n
```

```python
df_vgm2.drop('Unnamed: 0', axis = 1, inplace=True)
df_vgm2
```

```python
df_vgm2['NA_Sales'] = df_vgm2['NA_Sales'].str.replace(pat = r'[M K]', repl = r'', regex = True)
df_vgm2['EU_Sales'] = df_vgm2['EU_Sales'].str.replace(pat = r'[M K]', repl = r'', regex = True)
df_vgm2['JP_Sales'] = df_vgm2['JP_Sales'].str.replace(pat = r'[M K]', repl = r'', regex = True)
df_vgm2['Other_Sales'] = df_vgm2['Other_Sales'].str.replace(pat = r'[M K]', repl = r'', regex = True)

df_vgm2
```

```python
df_vgm2[['NA_Sales', 'EU_Sales','JP_Sales', 'Other_Sales']] =  df_vgm2[['NA_Sales', 'EU_Sales','JP_Sales', 'Other_Sales']].\
apply(pd.to_numeric)

df_vgm2.dtypes
```

```python
for i in index_n:
    df_vgm2['NA_Sales'][i] = df_vgm2['NA_Sales'][i] * 0.001
    print(df_vgm2['NA_Sales'][i])
```
